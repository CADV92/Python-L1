#### **Problema 1**
Solucin de una ecuacin de segundo grado
 $$\mathbf{a}x^{2} + \mathbf{b}x + \mathbf{c} ...... (1)$$

Discriminante de una ecuacin de segundo grado: ($\Delta$)

$$\Delta = \mathbf{b} ^{2} - 4\mathbf{a}\mathbf{c} ..... (2)$$

$$\mathbf{x_{1}} = \frac{- b + \sqrt{\Delta}}{2a}$$

$$\mathbf{x_{2}} = \frac{- b - \sqrt{\Delta}}{2a}$$

Donde:  $\mathbf{x_{1}}$ y $\mathbf{x_{2}}$, son las soluciones de la ecuacin  $\mathbf{(1)}$

**Condiciones**

*  si $\Delta > 0 $, existen dos soluciones
*  si $\Delta = 0 $, existen una nica solucin
*  si $\Delta < 0 $, existen no existen soluciones reales

```python
# SUMANDO MATRICES A y B
A = [[1, 4, 5], 
    [-5, 1, 9],
    [-6, 7, 1]]

B = [[5,4,7],
    [2,4,3],
    [9,6,1]]

C = [[0,0,0],
    [0,0,0],
    [0,0,0]]
```

```python
# Recorrido de los elementos de la matriz
for i in range(len(A)):                 # Iterando sobre las filas
    #print(f"fila {i+1} ---> ",A[i])
    for j in range(len(A[i])):          # Iterando sobre las columnas
        print(f"columna {i+1},{j+1} ---> ",A[i][j])
```

```python
# Recorrido de los elementos de la matriz
print()
for i in range(len(A)):                 # Iterando sobre las filas
    for j in range(len(A[i])):
       C[i][j] = A[i][j] + B[i][j]
    print(f"fila {i+1:02d} \t:",C[i])
```

```python
nfila = 3
ncolumna = 3

for i in range(nfila):
    for j in range(ncolumna):
        C[i][j] = A[i][j] + B[i][j]
    print(f"fila {i+1:08.4f} \t:",C[i])
```

```python
# Registro de datos
nombre = []
edad = []

while True:
    name = input("Ingrese el nombre: ")
    if name == "EOF":
        break
    age = input("Ingrese la edad: ")
    nombre.append(name)
    edad.append(age)

print("Nombres: \n\t", nombre)
print("Edad: \n\t", edad)
```
